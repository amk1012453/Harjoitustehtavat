```python
correctName = "Justin"
guesses = 0

while True:
    guess = input("Please, guess my name.\n")
    guesses += 1

    if guess.lower() == correctName.lower():
        print("Congratulations!")
        print("Guesses:", guesses)
        break

    else:
        quitting = input("Do you want to quit (y/n)?\n")
        if quitting.lower() == "y":
            print("Thank you for playing")
            break

```
